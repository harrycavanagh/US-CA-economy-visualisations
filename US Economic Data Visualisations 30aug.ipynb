```python
data2 = pd.read_csv("/Users/harry/Dropbox (Sandtable)/Root/Harry Workspace/US economic data/US economic data/economic data together.csv",header = None, usecols = cols,delimiter=',', encoding="utf-8-sig", nrows = 126)
date=data2[0].tolist()
inflation=data2[1].tolist()
pay=data2[2].tolist()
CPI=data2[3].tolist()
RGDP=data2[4].tolist()
ex_rate=data2[5].tolist()
unemployment=data2[6].tolist()
date.remove(date[0])
inflation.remove(inflation[0])
pay.remove(pay[0])
CPI.remove(CPI[0])
RGDP.remove(RGDP[0])
ex_rate.remove(ex_rate[0])
unemployment.remove(unemployment[0])
```

```python
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt


dates = []
for month in range(3, 13):
        dates.append(dt.datetime(year=2006, month=month, day=1))
for year in range(2007, 2016):
    for month in range(1, 13):
        dates.append(dt.datetime(year=year, month=month, day=1))
for month in range(1, 8):
        dates.append(dt.datetime(year=2016, month=month, day=1))
```

```python
inflation1 =[]
for i in inflation:
    inflation1.append(float(str(i))-float(str(inflation[0])))
pay1 =[]
for i in pay:
    pay1.append(float(str(i))-float(str(pay[0])))
CPI1 =[]
for i in CPI:
    CPI1.append(((float(str(i))-float(str(CPI[0])))/10))
RGDP1 =[]
for i in RGDP:
    RGDP1.append(float(str(i))-float(str(RGDP[0])))
ex_rate1 =[]
for i in ex_rate:
    ex_rate1.append(float(str(i))-float(str(ex_rate[0])))
unemployment1 =[]
for i in unemployment:
    unemployment1.append(float(str(i))-float(str(unemployment[0])))
```

```python
import matplotlib
%matplotlib inline
plt.figure(figsize=(30,20))
matplotlib.rc('xtick', labelsize=20) 
matplotlib.rc('ytick', labelsize=20) 
mpl.rcParams.update({'font.size': 22, 'font.family': 'serif'})
plt.ylabel('Variation in arbitrary units', fontsize = '30')
plt.plot(dates, inflation1, color = 'green', label = 'Inflation Rate')
plt.plot(dates, pay1, color = 'blue',label = 'Seasonally Adjusted Dollars per Hour')
plt.plot(dates, CPI1,color = 'purple',label = 'Consumer Price Index - FOOD & BEVERAGES')
plt.plot(dates, RGDP1,color = 'orange',label = 'Real gross domestic product (percent change from  \npreceding period,seasonally adjusted annual rate)')
plt.plot(dates, ex_rate1,color = 'black',label = 'GBP/USD exchange rate')
plt.plot(dates, unemployment1,color = 'red',label = 'Civillian Unemployment Rate (Percent,Seasonally Adjusted)')
plt.legend(loc=0,prop={'size':25})
plt.title(" US Economic Variables from 2006 - 2016", fontsize = 30,position=(0.5, 1.03))
plt.show()
```
