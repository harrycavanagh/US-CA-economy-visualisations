```python
import pandas as pd
import math as math
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl


pd.options.display.max_rows = 5


data = pd.read_csv("/Users/harry/Dropbox (Sandtable)/Root/Harry Workspace/US economic data/CA economic data/CA economic data.csv", delimiter=',', encoding="utf-8-sig")
print('California Economic Data')
print('DATA SOURCE: http://www.dot.ca.gov/')

data
```

```python
year=data['Year'].tolist()
employment=data['Employment percentage'].tolist()
pop=data['Population'].tolist()
pop=[ i.replace(",","") for i in pop ]
pop=[float(str(i))/10000000 for i in pop]

income=data['Per Capita Income (thousands)'].tolist()
ind_output=data['Industrial_Output'].tolist()
```

```python
employment1 = [(i-employment[0])*10 for i in employment]
pop1 =[(i-pop[0])*50 for i in pop]
income1 = [(i-income[0]) for i in income]
ind_output1 = [(i - ind_output[0])/10 for i in ind_output]
```

```python
%matplotlib inline

mpl.rcParams.update({'font.size': 22, 'font.family': 'serif'})
plt.figure(figsize=(30,20))

plt.ylabel('Variation in arbitrary units', fontsize = '30')
plt.xlabel('Year', fontsize = '30')

plt.plot(year[:17], employment1[:17], color = 'red', marker = 'o', label = 'Employment')
plt.plot(year[16:], employment1[16:], color = 'red', marker = 'o', linestyle = 'dashed')

plt.plot(year[:17], pop1[:17], color = 'blue', marker = 'o', label = 'Population')
plt.plot(year[16:], pop1[16:], color = 'blue', marker = 'o', linestyle = 'dashed')

plt.plot(year[:17], income1[:17], color = 'black', marker = 'o', label = 'Income')
plt.plot(year[16:], income1[16:], color = 'black', marker = 'o', linestyle = 'dashed')

plt.plot(year[:17], ind_output1[:17], color = 'purple', marker = 'o', label = 'Industrial Ouput')
plt.plot(year[16:], ind_output1[16:], color = 'purple', marker = 'o', linestyle = 'dashed')

plt.legend(loc=0,prop={'size':30})

plt.title("California Economic Variables 2000 - 2040", fontsize = 30,position=(0.5, 1.03))
plt.show()
```
